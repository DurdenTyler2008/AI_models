### Загружаю картинку самолета и читаю ее OpenCV:

```
im = cv2.imread('/content/drive/MyDrive/Самолеты_база/airplane_2.jpg')
 
newHeight = 200                               #высота
newWidth = int(im.shape[1]*200/im.shape[0])   #пропорционально этому меняею ширину
im = cv2.resize(im, (newWidth, newHeight))    #применяю это к изображению

%matplotlib inline
plt.imshow(im)
plt.show()
```

###Selective Search(селективный поиск):

```
cv2.setUseOptimized(True)     #режим оптимизации
cv2.setNumThreads(8)          #количество потоков
```

```
sss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation() #oбъект Selective Search Segmentation
sss.setBaseImage(im)
sss.switchToSelectiveSearchQuality()  #метод

rects = sss.process()            #результат вернётся в виде прямоугольников внутри которых предположительно находятся объекты
print(rects.shape)               #количество прямоугольников
```

```
numShowRects = 10                  #кол-во region proposals (прямоугольников, описывающих объект)

imOut = im.copy()

for i, rect in enumerate(rects): 
  if i < numShowRects:            #пока не наберётся numShowRects кол-во прямоугольников
      x, y, w, h = rect           # x,y - координаты точки вершины; w, h - шаги от вершины вправо, вниз
      cv2.rectangle(imOut, (x, y), (x+w, y+h), (0, 255, 0), 1, cv2.LINE_AA) #метод нарисует на изображении прямоугольник по указанным параметрам
      print("координаты:", "x", x, "y", y) 
  else: 
      break 

plt.imshow(imOut) 
plt.show()
```
